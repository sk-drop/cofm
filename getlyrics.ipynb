```python
import requests 
import os
import pandas as pd
import re
import json
from musixmatch import Musixmatch
```

```python
# re patterns for 1 artists and 2 titles
pattern = re.compile(r"^[\w.]+(?: (?!featuring\b)\w+)?")
pattern2 = re.compile(r".+?(?=\/)")
```

```python
# cutting searchstrings
def cutArtist(text):
    m = pattern.search(text)
    if m:
        return m.group(0)
    return text   

def cutTitle(text):
    m = pattern2.search(text)
    if m:
        return m.group(0)
    return text
```

```python
# get all lyrics
import lyricsgenius

df = pd.read_csv("./data/allSongs.csv")
lyricList = []
genius = lyricsgenius.Genius("mMzJ7AO3pIAnincdq0rplM5fje0sMrpabpqRZs0iC2_HStwfZDrNiiCa_uK-54ek", timeout=15)

for x, y in zip(df['Title'], df['Artist(s)']):
    try:
        lyrics = genius.search_song(cutTitle(x),cutArtist(y))
        lyrics = json.loads(lyrics.to_json())
        lyricList.append(lyrics['lyrics'])
    except:
        lyricList.append("none found")
        continue

df["Lyrics"] = lyricList
#df.to_csv("./data/lyrics/allLyrics.csv")
```
